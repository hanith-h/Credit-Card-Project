Detecting Missing Values
df.isnull() â€“ Checks for nulls

df.isnull().sum() â€“ Total nulls per column

df.info() â€“ Overview of nulls and data types

âœ… Methods to Handle Missing Values
1. Removing
Drop rows: df.dropna()

Drop columns: df.dropna(axis=1)

âœ… Use if missing values are few and data loss is minimal.

2. Imputation (Filling)
Mean: df['col'].fillna(df['col'].mean())

Median: df['col'].fillna(df['col'].median())

Mode: df['col'].fillna(df['col'].mode()[0])

Forward Fill: df.fillna(method='ffill')

Backward Fill: df.fillna(method='bfill')

Custom Value: df.fillna(0) or 'Unknown'

Interpolate (numerical): df.interpolate()

```python
#Handling Outliers
def handling_outliers(df):
    for col in df.select_dtypes(include='number').columns:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        lower_bound=Q1-(1.5*IQR)
        upper_bound=Q3+(1.5*IQR)
        #Capping
        df[col]=df[col].apply(lambda x:lower_bound if x<lower_bound else upper_bound
                              if x>upper_bound else x)
    return df
    
        
```

Scaling
-------

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x_scale=ss.fit_transform(x_selected)
```

```python
#Importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.2,random_state=40)
```

```python
#Printing the Shapes of the Sets
print(' x_train shape:',x_train.shape,'x_test shape:',x_test.shape,'y_train shape:',y_train.shape,'y_train shape:',y_test.shape)
```

```python
#Importing the ALGORITHM [Logistic Regression]
from sklearn.linear_model import LogisticRegression
```

```python
#Assigning LR to a variable
lr=LogisticRegression()
```

```python
#Training the model
lr.fit(x_train,y_train)
```

```python
#Predicting 
ypred_lr=lr.predict(x_test)
```

```python
#Importing MAE,MSE,R2
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
```

```python
#Evaluating accuracy_score,f1_score,precision_score,recall_score
accuracy=accuracy_score(y_test,ypred_lr)
precision=precision_score(y_test,ypred_lr)
recall=recall_score(y_test,ypred_lr)
f1=f1_score(y_test,ypred_lr)
print('ACCURACY:',accuracy)
print('PRECISION:',precision)
print('recall:',recall)
print('f1:',f1)
```
